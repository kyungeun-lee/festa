```python
from libs.data import TabularDataset
import xgboost as xgb
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def BinBoundaries(dataset, X_num, bin_size, device):
    init_num = torch.linspace(0, 1, bin_size + 1, device=device)
    init_bin_boundaries = dict()
    for f in range(dataset.size(1)):
        if f in X_num:
            init_bin_boundaries[f] = init_num
        else:
            init_bin_boundaries[f] = None
    return init_bin_boundaries
```

```python
from scipy.stats import wasserstein_distance
from scipy.spatial.distance import euclidean
from sklearn.metrics.pairwise import cosine_similarity

def calculate_emd(original, augmented):
    emd_values = [wasserstein_distance(original[:, i], augmented[:, i]) for i in range(original.shape[1])]
    return np.mean(emd_values)

def calculate_average_euclidean_distance(original, augmented):
    distances = [euclidean(original[i], augmented[i]) for i in range(len(original))]
    return np.mean(distances)

def calculate_cosine_similarity(original, augmented):
    similarities = [cosine_similarity(original[i].reshape(1, -1), augmented[i].reshape(1, -1))[0][0] for i in range(len(original))]
    return np.mean(similarities)

def get_results(transform_func, data, true_label):
    sample = transform_func({"image": torch.tensor(data), "mask": None})
    preds = model.predict(sample["image"].cpu().numpy())
    probs = model.predict_proba(sample["image"].cpu().numpy())
    acc, auc = evaluate(preds, true_label, prediction_proba=probs, tasktype="multiclass")
    was = calculate_emd(data, sample["image"].cpu().numpy())
    euc = calculate_average_euclidean_distance(data, sample["image"].cpu().numpy())
    cos = calculate_cosine_similarity(data, sample["image"].cpu().numpy())
    return acc, auc, was, euc, cos
```

```python
# results = pd.DataFrame(columns=("data_id", "numbins", "acc", "auroc", "cos", "euc", "was"))
# results.to_csv("stats.csv")
results = pd.read_csv("stats.csv", index_col=0)
print(results.shape)
```

```python
openml_id = 182
tasktype = "multiclass"
device = "cpu"

dataset = TabularDataset(openml_id, tasktype, device, labeled_data=0, cat_threshold=20, seed=1, modelname="xgboost")
(X_train, y_train), (X_val, y_val), (X_test, y_test) = dataset._indv_dataset()

if tasktype == "multiclass":
    print("Classes: ", y_train.size(1))
    y_train = np.argmax(y_train.cpu().numpy(), axis=1)
    y_val = np.argmax(y_val.cpu().numpy(), axis=1)
    y_test = np.argmax(y_test.cpu().numpy(), axis=1)
else:
    y_train = y_train.cpu().numpy()
    y_val = y_val.cpu().numpy()
    y_test = y_test.cpu().numpy()
print(X_train.size())
```

```python
X_train = X_train.cpu().numpy()
X_val = X_val.cpu().numpy()
X_test = X_test.cpu().numpy()

if tasktype == "multiclass":
    model = xgb.XGBClassifier(
            tree_method='auto', objective="multi:softmax",
            gpu_id=0, booster='gbtree', n_estimators=100,
            max_depth=3, subsample=1, eta=0.1, random_state=1
    ).fit(X_train, y_train)
else:
    model = xgb.XGBClassifier(
            tree_method='auto', objective="binary:logistic",
            gpu_id=0, booster='gbtree', n_estimators=100,
            max_depth=3, subsample=1, eta=0.1, random_state=1
    ).fit(X_train, y_train)
```

```python
from libs.transform import *
from libs.utils import evaluate
import pandas as pd
i = len(results)

# results = pd.DataFrame(columns=("data_id", "numbins", "acc", "auroc", "cos", "euc", "was"))
for b in np.arange(1, 20):
    bin_boundaries = BinBoundaries(torch.tensor(X_train), dataset.X_num, b, "cpu")
    binshuffling = BinShuffling(1, bin_boundaries, sampling_alpha=False)
    
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)
    acc, auc = evaluate(preds, y_test, prediction_proba=probs, tasktype="multiclass")

    acc3, auc3, was3, euc3, cos3 = get_results(binshuffling, X_test, y_test)
    
    results.loc[i] = [openml_id, b, acc3, auc3, cos3, euc3, was3]; i += 1
    
results.to_csv("stats.csv")
```

```python
from libs.transform import *
from libs.utils import evaluate
import pandas as pd
result = pd.DataFrame(columns=("p", "augmentation", "acc", "auc", "was", "euc", "cos"))
i = 0

for p in np.arange(0, 1.01, 0.01):
    masking = Masking(p, sampling_alpha=False)
    shuffling = Shuffling(p, sampling_alpha=False)

    bin_boundaries = BinBoundaries(torch.tensor(X_train), dataset.X_num, 20, "cpu")
    binshuffling = BinShuffling(p, bin_boundaries, sampling_alpha=False)
    
    preds = model.predict(X_test)
    probs = model.predict_proba(X_test)
    acc, auc = evaluate(preds, y_test, prediction_proba=probs, tasktype="multiclass")

    acc1, auc1, was1, euc1, cos1 = get_results(masking, X_test, y_test)
    acc2, auc2, was2, euc2, cos2 = get_results(shuffling, X_test, y_test)
    acc3, auc3, was3, euc3, cos3 = get_results(binshuffling, X_test, y_test)
    
    result.loc[i] = [p, "masking", acc1, auc1, was1, euc1, cos1]; i += 1
    result.loc[i] = [p, "shuffling", acc2, auc2, was2, euc2, cos2]; i += 1
    result.loc[i] = [p, "binshuffling", acc3, auc3, was3, euc3, cos3]; i += 1
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))
for l in ["masking", "shuffling", "binshuffling"]:
    q = result[result["augmentation"] == l]
#     plt.plot(q["p"].values, q["acc"].values, label=l.capitalize())
    plt.plot(q["p"].values, q["cos"].values, label=l.capitalize())
plt.rc('axes', axisbelow=True)
plt.grid()
# plt.hlines(y=acc, xmin=-0.5, xmax=1.5, color="red", linewidth=0.8, linestyle="dashed", label="Original data")
plt.xlim(0, 1)
# plt.yscale("log")
plt.xlabel("Masking probability")
plt.ylabel("Euclidean distance (log scale)")
plt.legend()
plt.show()
```
