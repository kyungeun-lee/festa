```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
datalist = [
    22, 54, 1063, 1067, 12, 18, 23, 59, 188, 307, 1043, 1459, 1475, 1489, 1492, 1497, 1503, 4153, 40499, 44125,
    44131, 45062, 44157, 1462, 44160, 29, 37, 53, 49, 1504, 1494, 41143, 44126, 40981, 41168, 44091, 44158, 44123, 44090,
    40922, 44161, 45714]
```

```python
result = pd.read_csv("/home/SemiTab/result-summary/result.csv", index_col=0)
result = result[result["model"].str.startswith("sslbins")].reset_index(drop=True)
result = result[["trial", "shots", "data_id", "model", "acc"]]

i = len(result)
for trial in range(10):
    for model in ["sslsmallnoise", "ssloriginal"]:
        for shot in [1, 5]:
            print("")
            print("#####", trial, model, shot)
            for d in datalist:
                fname = f'/home/SemiTab/results/seed={trial}/shot={shot}/model={model}/data={d}/performance.npy'
                if os.path.exists(fname):
                    try:
                        perf = np.load(fname, allow_pickle=True)
                        perf = perf.item()
                        result.loc[i] = [trial, shot, d, model, perf["Test"]["lr"][0]]
                        i += 1
                    except Exception:
                        pass
    #                     print(d, end=" ")
                else:
                    print(d, end=" ")
                    pass
```

```python
result = result[(result["trial"] < 10) & (
    result["model"].isin(["sslbinsampling-lr", "sslbinshuffling-lr", "ssloriginal", "sslsmallnoise"]))].reset_index(drop=True)
result.groupby(["shots", "model"]).mean("acc").reset_index()
```

```python
result.groupby(["shots", "model"]).mean("acc").reset_index().to_csv("result-summary/")
```
